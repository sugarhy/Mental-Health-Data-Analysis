# Introduction 

Mental health in university students is a significant concern as the transition to college life can bring about various challenges that can affect their well-being. Students frequently struggle with anxiety, despair, and panic attacks. Academic demands, social adjustments, financial worries, marital issues, and the general environment change can all contribute to mental health problems. <br>

Hence, it is extremely essential for universities to prioritise mental health initiatives and create a comprehensive support system that addresses the diverse needs of students. <br>

I will be analysing the mental health of univerisity students in this notebook to identify the attributes that leads to poor mental health. <br>

The 3 mental health disorders that will be discussed here are depression, anxiety and panic attacks.

<b> 1. Handle Missing Values </b>

```python
# determine column that has the missing value in the Age column 
missing_entry = pd.isnull(data['Age'])

# display row of data where Age is NULL 
data[missing_entry]
```

```python
# drop NA value 
data = data[data['Age'].notna()]
data.shape
```

```python
data.head(1)
```

<b> 2. Standardise Data Names </b>

a. Simplify column names to a more standard convention 

```python
# rename column names
data.columns = ['Timestamp', 'Gender', 'Age', 'Course', 'Year', 'CGPA', 'Marital_Status', 'Depression', 'Anxiety', 'Panic_Attack', 'Treatment']
data.head(1)
```

b. Standardise input values

```python
#Formatting the year column 
data['Year'].unique()
```

Findings from the above cell:
* College years range from Year 1 to Year 4
* year 1 and Year 1 mean the same thing 
* We will change the data type for the year to INT such that the year levels are represented with just numbers (e.g year 1 -> 1)


```python
# function to change Year to INT datatype
def text_to_int(Text):
    Text = Text[-1] #last element(the year number)
    Text = int(Text)
    return Text

data["Year"] = data["Year"].apply(text_to_int)
print("First three values of Year after cleaning text:")
print(data["Year"][:3], "\n")
```

<b>3. Remove Unwanted Observations (e.g duplicates)</b>

```python
data['CGPA'].unique()
```

Findings:
* The CGPA column has ranges rather than an absolute value
* There are 2 repeated ranges '3.50 - 4.00' (no whitespace at the end) and '3.50 - 4.00 '(has whitespace at the end)

```python
# function to remove the whitespace
# using .strip() to remove the space
def remove_space(string):
    string = string.strip()
    return string 

data["CGPA"] = data["CGPA"].apply(remove_space)

print("First 3 values of CGPA after cleaning text:")
print(data["CGPA"][:3], "\n")
print(data["CGPA"].unique())
```

```python
data["Course"].unique()
```

Findings:<br>
* We can observe that a lot of courses are interpreted differently though they refer to the same course. For example, the courses "Engine", "engin" and "Engineering" all refer to the Engineering course. We clean and standardise all the course names. 

```python
data["Course"].replace({"engin":"Engineering", "Engine": "Engineering", 
                        "Islamic education": "Islamic Education", 
                        "BIT": "IT",
                        "Laws": "Law",
                        "Mathemathics":"Mathematics", 
                       "Pendidikan islam": "Pendidikan Islam",
                       "Pendidikan Islam ": "Pendidikan Islam",
                       "Usuluddin ": "Usuluddin",
                       "koe": "KOE",
                        "Benl": "BENL",
                       "Econs": "Economics",
                       "Human Sciences ":"Human Sciences",
                       "Communication ": "Communication",
                        "KIRKHS": "Irkhs",
                        "Kirkhs": "Irkhs",
                       "psychology": "Psychology",
                        "Fiqh fatwa": "Fiqh",
                        "Nursing ": "Nursing"}, inplace=True)

len(data["Course"].unique())
```

# Data Preprocessing

#### Label Encoding 

=> a technique used in machine learning and data analysis to convert categorical variables into numerical format. It is useful when working with algorithms that require numerical input, as most machine learning models can only operate on numerical data.

=> Each label is assigned a unique integer based on alphabetical ordering 


```python
# drop the 'timestamp' column 
data.drop('Timestamp', axis=1, inplace=True)

data.head()
```

```python
# label encoding 
from sklearn.preprocessing import LabelEncoder 

# LabelEncoder() object knows how to understand word labels
encoder = LabelEncoder()

# identify all the columns with categorical datatypes and group them
# gender, marital status, depression, anxiety, panic attack, treatment
categorical_columns =[x for x in data.columns if data.dtypes[x] == 'object']

# encode labels for all categorical columns
for column in categorical_columns:
    data[column] = encoder.fit_transform(data[column])
    
data.head()
```

```python
# correlation matrix 
corrmat = data.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corrmat, annot=True, cmap="crest")
```

Key Findings:
* Marital status and depression have the closest association of 0.59.  
* Depression, anxiety and panic attacks are relatively associated with one another 

# Model Selection 

For model selection, I will be building pipelines of five different classifiers and select one with the best fit results. 
In this section:
1. split data into training and testing sets
2. assign targets and features 
3. the model pipelines with preprocessing 
4. fit the training set to the various models 
5. get the confusion matrix and accuracy score
6. pick the best classifier    

```python
X = data.drop(["CGPA"], axis=1)
y = data["CGPA"]
```

```python
# splitting test and training sets
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.3, 
                                                    random_state=42)
```

```python
# model pipelines

# Logistic Regression 
pipeline_lr=Pipeline([('lr_classifier', LogisticRegression(random_state=42))])

# Decision Tree Classifier
pipeline_dt=Pipeline([('dt_classifier', DecisionTreeClassifier(random_state=42))])

# Random Forest Classifier
pipeline_rf=Pipeline([('rf_classifer', RandomForestClassifier())])

# SVC 
pipeline_svc=Pipeline([('sv_classifier', SVC())])


#List of all the pipelines 
pipelines = [pipeline_lr, pipeline_dt, pipeline_rf, pipeline_svc]

#Dictionary of pipelines and classifier types for ease of reference
pipe_dict={0:'Logistic Regression', 1:'Decision Tree', 2:'RandomForest', 3:'SVC'}

# Train the model using the training datasets 
for pipe in pipelines:
    pipe.fit(X_train, y_train)
    
#cross validation on accuracy 
cv_results_accuracy=[]
for i, model in enumerate(pipelines):
    cv_score = cross_val_score(model, X_train, y_train, cv=10)
    cv_results_accuracy.append(cv_score)
    print("%s: %f " % (pipe_dict[i], cv_score.mean()))
```

We can see that Random Forest results in the most accurate model. 

# Model Evaluation

1. Build a Random Forest Model 
2. Create a classification report 
3. Visualise the confusion matrix

```python
pred_rfc = pipeline_rf.predict(X_test)
accuracy = accuracy_score(y_test, pred_rfc)
print(accuracy)
```

```python
RF_model = RandomForestClassifier()
RF_model.fit(X_train, y_train)

# testing the model on test set
predictions = RF_model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
accuracy
```

```python
accuracy = accuracy_score(y_test, predictions)
recall = recall_score(y_test, predictions, average="weighted")
precision = precision_score(y_test, predictions, average="weighted")
f1_score = f1_score(y_test, predictions, average="micro")


print("********* Random Forest Results *********")
print("Accuracy    : ", accuracy)
print("Recall      : ", recall)
print("Precision   : ", precision)
print("F1 Score    : ", f1_score)
```

```python
print(classification_report(y_test, predictions))
```

```python
plt.subplots(figsize=(12,8))
cf_matrix = confusion_matrix(y_test, predictions)
sns.heatmap(cf_matrix/np.sum(cf_matrix), cmap=None, annot=True, annot_kws={"size": 15})
```
